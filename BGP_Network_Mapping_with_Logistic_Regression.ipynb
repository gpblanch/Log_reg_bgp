```python
#!/usr/bin/env python3
"""
BGP Network Mapping with Logistic Regression
Complete implementation for analyzing AS relationships using real and synthetic network data.
"""

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve
import requests
import bz2
import gzip
import io
import os
from datetime import datetime
import warnings
from IPython.display import display # For Colab-friendly DataFrame display

warnings.filterwarnings('ignore')

# --- Configuration ---
# IMPORTANT: For reliable BGP processing, download a RIB file and set this path.
# Example: "/path/to/your/downloaded/rib.20240101.0000.bz2"
USER_PROVIDED_RIB_FILE_PATH = None  # SET THIS TO YOUR LOCAL RIB FILE PATH
# If USER_PROVIDED_RIB_FILE_PATH is None, the script will attempt to download an example file.
EXAMPLE_RIB_URL = "http://archive.routeviews.org/bgpdata/2025.05/RIBS/rib.20250501.0000.bz2"
EXAMPLE_RIB_FILENAME = "rib.20250501.0000.bz2" # Expected filename for the example

# Attempt to import bgpkit-parser
try:
    from bgpkit import Parser
    BGPKIT_AVAILABLE = True
    print("bgpkit-parser is available.")
except ImportError:
    BGPKIT_AVAILABLE = False
    print("WARNING: bgpkit-parser not installed. Real BGP feature extraction will be largely SKIPPED.")
    print("The script will rely on synthetic data for BGP features.")
    print("To enable BGP feature extraction, please install it with: pip install bgpkit-parser")

# Set up plotting style
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

print("=" * 80)
print("BGP Network Mapping Analysis - AS Relationship Classification")
print("=" * 80)

# ========== SECTION 1: DATA RETRIEVAL (CAIDA AS Relationships) ==========

def download_caida_as_relationships():
    """
    Download CAIDA AS Relationships dataset.
    """
    print("\n1. DOWNLOADING CAIDA AS RELATIONSHIPS DATA")
    print("-" * 50)
    # Using a specific, likely stable link for CAIDA data
    # Using a known past date for better reliability for this example script.
    #caida_url = "https://publicdata.caida.org/datasets/as-relationships/serial-2/20240101.as-rel2.txt.bz2"
    caida_url = "https://publicdata.caida.org/datasets/as-relationships/serial-2/20250501.as-rel2.txt.bz2"

    try:
        print(f"Downloading AS relationship data from: {caida_url}")
        response = requests.get(caida_url, stream=True, timeout=30)
        response.raise_for_status()
        with bz2.BZ2File(filename=io.BytesIO(response.content)) as bz:
            content = bz.read().decode('utf-8')
        print("Successfully downloaded CAIDA AS relationship data.")
        return content
    except Exception as e:
        print(f"Error downloading CAIDA data: {e}")
        print("Falling back to generating sample AS relationship data.")
        return None # Triggers sample data generation

def parse_as_relationships(content):
    """
    Parse AS relationship data.
    Format: FromAS|ToAS|Relationship (-1 = c2p, 0 = p2p)
    """
    if content is None:
        print("Generating sample AS relationship data (CAIDA download failed or skipped).")
        return generate_realistic_as_data(type='caida')

    lines = content.strip().split('\n')
    data = []
    for line in lines:
        if line.startswith('#') or not line.strip():
            continue
        parts = line.split('|')
        if len(parts) >= 3:
            try:
                as1 = int(parts[0])
                as2 = int(parts[1])
                rel_type_raw = int(parts[2])
                # Map CAIDA format: -1 (customer-provider) to 1, 0 (peer-to-peer) to 0
                rel_type_mapped = 1 if rel_type_raw == -1 else (0 if rel_type_raw == 0 else -99) # -99 for other types
                if rel_type_mapped != -99: # Only include p2p and c2p
                     data.append([as1, as2, rel_type_mapped])
            except ValueError:
                continue
    df = pd.DataFrame(data, columns=['as1', 'as2', 'relationship_type'])
    if df.empty:
        print("CAIDA parsing resulted in an empty DataFrame. Generating sample data.")
        return generate_realistic_as_data(type='caida')
    return df

def generate_realistic_as_data(n_samples=5000, type='caida'):
    """
    Generate realistic placeholder data if real data download/parsing fails.
    'caida' type generates as1, as2, relationship_type.
    'bgp_features' type generates the full feature set synthetically.
    """
    np.random.seed(42)
    if type == 'caida':
        data = []
        for _ in range(n_samples):
            as1 = np.random.randint(1, 60000)
            as2 = np.random.randint(1, 60000)
            while as1 == as2: as2 = np.random.randint(1, 60000)
            rel_type = np.random.choice([0, 1], p=[0.4, 0.6]) # 0 for P2P, 1 for C2P
            data.append([as1, as2, rel_type])
        return pd.DataFrame(data, columns=['as1', 'as2', 'relationship_type']).drop_duplicates(subset=['as1', 'as2'])
    elif type == 'bgp_features': # For full synthetic fallback
        df_synth_caida = generate_realistic_as_data(n_samples=n_samples, type='caida')
        return _generate_synthetic_bgp_features(df_synth_caida)
    return pd.DataFrame()


# ========== SECTION 1B: DATA RETRIEVAL (RouteViews BGP Data) ==========

def download_example_routeviews_rib():
    """Downloads the example RouteViews RIB file if not already present."""
    if os.path.exists(EXAMPLE_RIB_FILENAME):
        print(f"Example RIB file '{EXAMPLE_RIB_FILENAME}' already exists locally.")
        return EXAMPLE_RIB_FILENAME

    print(f"\nAttempting to download example RIB file: {EXAMPLE_RIB_URL}")
    print("This may take a while...")
    try:
        response = requests.get(EXAMPLE_RIB_URL, stream=True, timeout=300) # Increased timeout for large file
        response.raise_for_status()
        with open(EXAMPLE_RIB_FILENAME, 'wb') as f:
            for chunk in response.iter_content(chunk_size=8192 * 10): # Larger chunk size
                f.write(chunk)
        print(f"Successfully downloaded '{EXAMPLE_RIB_FILENAME}'.")
        return EXAMPLE_RIB_FILENAME
    except requests.exceptions.RequestException as e:
        print(f"Failed to download example RIB file '{EXAMPLE_RIB_URL}': {e}")
        return None
    except Exception as e:
        print(f"An unexpected error occurred during RIB download: {e}")
        return None

# ========== SECTION 2: BGP FEATURE ENRICHMENT ==========

def _generate_synthetic_bgp_features(df_input):
    """
    Generates a full set of SYNTHETIC BGP features.
    This is used as a fallback if real BGP processing fails or is skipped.
    """
    print("Generating SYNTHETIC BGP features as a fallback.")
    n = len(df_input)
    if n == 0: return df_input # Return empty if input is empty

    features = df_input.copy()

    # CAIDA-based degree (already calculated if df_input is from enrich_with_bgp_features)
    if 'as1_degree_caida' not in features.columns:
        caida_as_degrees = {}
        all_asns_caida = pd.concat([features['as1'], features['as2']]).unique()
        adj_list_caida = {asn: [] for asn in all_asns_caida}
        for _, row in features.iterrows():
            adj_list_caida[row['as1']].append(row['as2'])
            adj_list_caida[row['as2']].append(row['as1'])
        caida_as_degrees = {asn: len(neighbors) for asn, neighbors in adj_list_caida.items()}
        features['as1_degree'] = features['as1'].map(lambda x: caida_as_degrees.get(x, np.random.randint(1,5)))
        features['as2_degree'] = features['as2'].map(lambda x: caida_as_degrees.get(x, np.random.randint(1,5)))
    else: # if CAIDA degrees were pre-calculated
        features['as1_degree'] = features['as1_degree_caida']
        features['as2_degree'] = features['as2_degree_caida']


    # Synthetic versions of other BGP features
    features['avg_path_length_diff'] = np.random.normal(1.0, 1.0, n) * (2 * features['relationship_type'] - 1) # C2P tend to have diff > 0
    features['prefix_ratio'] = np.random.lognormal(mean=np.log(1.0 + features['relationship_type']), sigma=0.5, size=n) # C2P higher ratio
    features['geographic_distance'] = np.random.exponential(scale=1000, size=n) + 50 * (1 - features['relationship_type']) # P2P slightly closer
    features['common_neighbors'] = np.random.poisson(lam=2 + 3 * (1 - features['relationship_type']), size=n) # P2P more common_neighbors
    features['community_similarity'] = np.random.beta(a=2 + 3 * (1 - features['relationship_type']), b=5 - 3 * (1-features['relationship_type']), size=n) # P2P more similar
    features['route_stability'] = np.random.beta(a=7 - 2 * features['relationship_type'], b=3, size=n) # C2P slightly more stable
    features['prepending_freq'] = np.random.beta(a=2 + features['relationship_type'], b=7, size=n) # C2P might see more from customer
    features['path_diversity'] = np.random.gamma(shape=3 + (1-features['relationship_type'])*2, scale=1.5, size=n) # P2P more diverse

    # Noise (optional, applied if this function is the final step for features)
    numeric_cols = features.select_dtypes(include=[np.number]).columns
    numeric_cols = [col for col in numeric_cols if col not in ['as1', 'as2', 'relationship_type']]
    for col in numeric_cols:
        if features[col].std() > 0 and not np.isnan(features[col].std()):
            noise = np.random.normal(0, 0.1 * features[col].std(), n)
            features[col] += noise
        if col not in ['avg_path_length_diff']: features[col] = features[col].clip(lower=0)

    return features


def enrich_with_bgp_features(df_relationships, rib_file_path_to_process=None):
    print("\n2. ENRICHING DATA WITH BGP FEATURES")
    print("-" * 50)

    if df_relationships.empty:
        print("Input CAIDA DataFrame for feature enrichment is empty. Cannot proceed.")
        return pd.DataFrame() # Return empty DataFrame matching expected output structure later

    base_features_df = df_relationships.copy()
    n = len(base_features_df)

    # --- Initialize feature columns ---
    # CAIDA-based degree
    caida_as_degrees = {}
    all_asns_caida = pd.concat([base_features_df['as1'], base_features_df['as2']]).unique()
    adj_list_caida = {asn: set() for asn in all_asns_caida} # Use set for unique neighbors
    for _, row in base_features_df.iterrows():
        adj_list_caida[row['as1']].add(row['as2'])
        adj_list_caida[row['as2']].add(row['as1'])
    caida_as_degrees = {asn: len(neighbors) for asn, neighbors in adj_list_caida.items()}
    base_features_df['as1_degree_caida'] = base_features_df['as1'].map(caida_as_degrees).fillna(0)
    base_features_df['as2_degree_caida'] = base_features_df['as2'].map(caida_as_degrees).fillna(0)

    # Columns for BGP-derived features, initialized
    base_features_df['as1_degree_bgp'] = 0
    base_features_df['as2_degree_bgp'] = 0
    base_features_df['prefix_count_as1'] = 0
    base_features_df['prefix_count_as2'] = 0
    base_features_df['prefix_ratio'] = np.nan
    base_features_df['prepending_freq_as1'] = 0.0 # Simplified: ratio of prepended paths from AS1
    base_features_df['prepending_freq_as2'] = 0.0 # Simplified: ratio of prepended paths from AS2
    base_features_df['common_neighbors_bgp'] = 0

    # --- Attempt Real BGP Processing ---
    bgp_processing_successful = False
    if BGPKIT_AVAILABLE and rib_file_path_to_process and os.path.exists(rib_file_path_to_process):
        print(f"Attempting to process BGP RIB file: {rib_file_path_to_process} using bgpkit-parser.")
        print("This can be time-consuming for large RIB files...")

        as_prefix_counts_bgp = {}
        as_adjacencies_bgp = {asn: set() for asn in all_asns_caida} # Initialize for all ASNs in CAIDA set
        as_path_origins_bgp = {asn: 0 for asn in all_asns_caida} # Total paths originated
        as_prepending_incidents_bgp = {asn: 0 for asn in all_asns_caida} # Count of paths where ASN might be prepending

        try:
            parser = Parser(url=rib_file_path_to_process)
            processed_count = 0
            for elem in parser:
                processed_count += 1
                if processed_count % 500000 == 0: # Log progress
                    print(f"  Processed {processed_count} BGP records...")
                if elem.get("type") != "TABLE_DUMP_V2":
                   continue

                # Subtype can be RIB_IPV4_UNICAST, RIB_IPV6_UNICAST etc.
                # For simplicity, we'll count prefixes regardless of AF/SAFI for now
                # In a real scenario, you might filter by AFI/SAFI (e.g. elem.data.afi.name == "IPV4")

                origin_as_list = elem.get("origin_asns", [])
                if origin_as_list:
                    origin_as = origin_as_list[0]
                    as_prefix_counts_bgp[origin_as] = as_prefix_counts_bgp.get(origin_as, 0) + 1
                    as_path_origins_bgp[origin_as] = as_path_origins_bgp.get(origin_as, 0) + 1

                    # BGP Adjacencies and Prepending Detection
                    full_as_path = [int(a) for a in elem.get("as_path", "").split()]
                    if not full_as_path:
                        continue

                        # Prepending (simplified: if an AS appears multiple times consecutively or path is longer than unique ASNs)
                        # A more robust check would look at an AS prepending itself.
                        # This simple check looks if the origin AS is prepending in its own originated path
                        if len(full_as_path) > len(set(full_as_path)) and origin_as in full_as_path:
                             # Check if origin_as itself is part of the prepending.
                             # Example: 701 701 701 1234. Here 701 is prepending if it's origin_as.
                            idx = 0
                            while idx < len(full_as_path):
                                current_as = full_as_path[idx]
                                count = 1
                                idx_ahead = idx + 1
                                while idx_ahead < len(full_as_path) and full_as_path[idx_ahead] == current_as:
                                    count += 1
                                    idx_ahead += 1
                                if count > 1 and current_as == origin_as : # If origin AS repeated itself
                                    as_prepending_incidents_bgp[origin_as] = as_prepending_incidents_bgp.get(origin_as, 0) + 1
                                    break # Count one incident per path for this AS
                                idx = idx_ahead



                        # BGP Adjacencies
                        unique_asns_in_path = []
                        if full_as_path: # Create unique path, preserving order for adjacency
                            last_seen = None
                            for asn_in_path in full_as_path:
                                if asn_in_path != last_seen:
                                    unique_asns_in_path.append(asn_in_path)
                                    last_seen = asn_in_path

                        for i in range(len(unique_asns_in_path) - 1):
                            u, v = unique_asns_in_path[i], unique_asns_in_path[i+1]
                            if u is not None and v is not None and u != v: # Avoid self-loops here
                                if u not in as_adjacencies_bgp: as_adjacencies_bgp[u] = set()
                                if v not in as_adjacencies_bgp: as_adjacencies_bgp[v] = set()
                                as_adjacencies_bgp[u].add(v)
                                as_adjacencies_bgp[v].add(u)

            print(f"Finished BGP parsing. Processed {processed_count} records.")

            # Map BGP derived stats back to the DataFrame
            base_features_df['as1_degree_bgp'] = base_features_df['as1'].map(lambda x: len(as_adjacencies_bgp.get(x, set()))).fillna(0)
            base_features_df['as2_degree_bgp'] = base_features_df['as2'].map(lambda x: len(as_adjacencies_bgp.get(x, set()))).fillna(0)

            base_features_df['prefix_count_as1'] = base_features_df['as1'].map(as_prefix_counts_bgp).fillna(0)
            base_features_df['prefix_count_as2'] = base_features_df['as2'].map(as_prefix_counts_bgp).fillna(0)
            base_features_df['prefix_ratio'] = np.where(
                base_features_df['prefix_count_as2'] > 0,
                base_features_df['prefix_count_as1'] / base_features_df['prefix_count_as2'],
                np.nan # Or a large number if as1 has prefixes and as2 has 0
            )
            base_features_df['prefix_ratio'].fillna(base_features_df['prefix_count_as1'].apply(lambda x: x if x > 0 else 0), inplace=True) # Handle as2 having 0 prefixes

            # Prepending frequency (ratio of prepended paths to total originated paths)
            total_paths_as1 = base_features_df['as1'].map(as_path_origins_bgp).fillna(0)
            incidents_as1 = base_features_df['as1'].map(as_prepending_incidents_bgp).fillna(0)
            base_features_df['prepending_freq_as1'] = np.where(total_paths_as1 > 0, incidents_as1 / total_paths_as1, 0.0)

            total_paths_as2 = base_features_df['as2'].map(as_path_origins_bgp).fillna(0)
            incidents_as2 = base_features_df['as2'].map(as_prepending_incidents_bgp).fillna(0)
            base_features_df['prepending_freq_as2'] = np.where(total_paths_as2 > 0, incidents_as2 / total_paths_as2, 0.0)
            # Combine prepending frequency for the pair (e.g., max, or average, or specific to relationship direction)
            # For now, we use as1's prepending freq as the feature for the pair
            base_features_df['prepending_freq'] = base_features_df['prepending_freq_as1']


            # Common Neighbors based on BGP adjacencies
            common_neigh_list = []
            for _, row in base_features_df.iterrows():
                neighbors_as1 = as_adjacencies_bgp.get(row['as1'], set())
                neighbors_as2 = as_adjacencies_bgp.get(row['as2'], set())
                common_neigh_list.append(len(neighbors_as1.intersection(neighbors_as2)))
            base_features_df['common_neighbors_bgp'] = common_neigh_list

            bgp_processing_successful = True
            print("Successfully populated some features from BGP data: degree, prefix counts/ratio, basic prepending, common neighbors.")

        except Exception as e:
            print(f"Error during BGP file processing with bgpkit-parser: {e}")
            bgp_processing_successful = False

    else:
        if not BGPKIT_AVAILABLE:
            print("bgpkit-parser not found. Skipping real BGP processing.")
        if not rib_file_path_to_process or not os.path.exists(rib_file_path_to_process):
            print(f"RIB file not provided or not found at '{rib_file_path_to_process}'. Skipping real BGP processing.")

    # --- Consolidate and Generate/Fallback for other features ---
    final_df = base_features_df.copy()

    if bgp_processing_successful:
        final_df['as1_degree'] = final_df['as1_degree_bgp']
        final_df['as2_degree'] = final_df['as2_degree_bgp']
        final_df['common_neighbors'] = final_df['common_neighbors_bgp']
        # prefix_ratio and prepending_freq are already named as desired
        print("Using BGP-derived degrees, prefix_ratio, prepending_freq, common_neighbors.")
    else: # Fallback to CAIDA degrees if BGP degrees failed
        final_df['as1_degree'] = final_df['as1_degree_caida']
        final_df['as2_degree'] = final_df['as2_degree_caida']
        # For other features, if BGP processing failed, they'll need synthetic versions
        print("BGP processing failed or was skipped. Using CAIDA degrees and synthetic data for other BGP-specific features.")
        final_df['prefix_ratio'] = np.random.lognormal(mean=np.log(1.0 + final_df['relationship_type']), sigma=0.5, size=n)
        final_df['prepending_freq'] = np.random.beta(a=2 + final_df['relationship_type'], b=7, size=n)
        final_df['common_neighbors'] = np.random.poisson(lam=2 + 3 * (1 - final_df['relationship_type']), size=n)


    # Features that are generally synthetic or require more complex logic / external data
    print("Generating/Synthesizing remaining features (avg_path_length, geo_dist, community, stability, path_diversity).")
    final_df['avg_path_length_diff'] = np.random.normal(1.0, 1.0, n) * (2*final_df['relationship_type']-1) + \
                                   np.random.normal(0,0.5,n) # Centered around type, with noise
    final_df['geographic_distance'] = np.random.exponential(scale=1000, size=n) + 50 * (1 - final_df['relationship_type']) # P2P slightly closer
    final_df['community_similarity'] = np.random.beta(a=2 + 4 * (1 - final_df['relationship_type']), b=5 - 2 * (1-final_df['relationship_type']), size=n)
    final_df['route_stability'] = np.random.beta(a=7 - 2 * final_df['relationship_type'], b=3, size=n)
    final_df['path_diversity'] = np.random.gamma(shape=3 + (1-final_df['relationship_type'])*2, scale=1.5, size=n)

    # Select and order final columns
    # Ensure all features used by the model later are present
    feature_cols_final = [
        'as1_degree', 'as2_degree', 'avg_path_length_diff', 'prefix_ratio',
        'geographic_distance', 'common_neighbors', 'community_similarity',
        'route_stability', 'prepending_freq', 'path_diversity'
    ]
    # Add core columns
    all_cols = ['as1', 'as2', 'relationship_type'] + feature_cols_final

    # Ensure all these columns exist, fill with 0 or NaN if any are missing from processing steps
    for col in all_cols:
        if col not in final_df.columns:
            print(f"Warning: Column '{col}' was missing. Adding it with default values (0 or NaN).")
            final_df[col] = 0 # Or np.nan depending on desired handling

    # Prune to only necessary columns to avoid issues with extra intermediate columns
    final_df = final_df[all_cols]


    print(f"Feature enrichment complete. Final DataFrame shape: {final_df.shape}")
    return final_df


# ========== MAIN SCRIPT LOGIC ==========
if __name__ == '__main__':
    print(f"Script started at: {datetime.now()}")

    # --- 1. Get CAIDA AS Relationships ---
    as_data_content = download_caida_as_relationships()
    df_relationships = parse_as_relationships(as_data_content)

    if df_relationships.empty:
        print("CRITICAL: No AS relationship data (CAIDA or sample) could be loaded. Exiting.")
        exit()
    print(f"Loaded {len(df_relationships)} AS relationships from CAIDA/sample.")
    display(df_relationships.head())


    # --- 2. Prepare RIB file for BGP feature enrichment ---
    rib_file_to_use = None
    if USER_PROVIDED_RIB_FILE_PATH and os.path.exists(USER_PROVIDED_RIB_FILE_PATH):
        print(f"Using user-provided RIB file: {USER_PROVIDED_RIB_FILE_PATH}")
        rib_file_to_use = USER_PROVIDED_RIB_FILE_PATH
    else:
        if USER_PROVIDED_RIB_FILE_PATH: # Path was given but not found
             print(f"Warning: User-provided RIB file '{USER_PROVIDED_RIB_FILE_PATH}' not found.")
        print("Attempting to use the example RIB file.")
        rib_file_to_use = download_example_routeviews_rib() # Downloads if not present
        if not (rib_file_to_use and os.path.exists(rib_file_to_use)):
            print("No valid RIB file available for BGP feature processing. Will rely on synthetic BGP features.")
            rib_file_to_use = None # Ensure it's None if download failed

    # --- 3. Enrich features ---
    df = enrich_with_bgp_features(df_relationships, rib_file_path_to_process=rib_file_to_use)

    if df.empty:
        print("CRITICAL: DataFrame is empty after feature enrichment. Exiting.")
        exit()

    # Sample if dataset is too large for quick processing
    MAX_SAMPLES_FOR_EDA = 100000
    if len(df) > MAX_SAMPLES_FOR_EDA:
        print(f"\nSampling {MAX_SAMPLES_FOR_EDA} records from {len(df)} total for faster EDA and modeling...")
        df = df.sample(n=MAX_SAMPLES_FOR_EDA, random_state=42)

    print(f"\nFinal dataset shape for analysis: {df.shape}")
    print(f"Columns: {df.columns.tolist()}")
    print("\nPreview of the final dataset for analysis:")
    display(df.head())

    # ========== SECTION 3: EXPLORATORY DATA ANALYSIS ==========
    print("\n3. EXPLORATORY DATA ANALYSIS")
    print("-" * 50)

    print("\nTarget Variable Distribution:")
    display(df['relationship_type'].value_counts().to_frame(name='Count'))
    print(f"\nPercentage distribution:")
    display((df['relationship_type'].value_counts(normalize=True) * 100).to_frame(name='Percentage'))

    print("\nMissing values per column:")
    display(df.isnull().sum().to_frame(name='Missing Values').T) # Transpose for better single-row display

    # Feature columns for analysis (ensure these are the ones in the final df)
    feature_cols = [col for col in df.columns if col not in ['as1', 'as2', 'relationship_type']]

    print("\nBasic Statistics of Features:")
    display(df[feature_cols].describe())

    # ========== SECTION 4: VISUALIZATIONS ==========
    # Tooltip descriptions for these plots were provided separately.
    # The code below generates static Matplotlib plots.

    print("\nGenerating visualizations...")
    # Ensure feature_cols are all numeric for plotting certain types
    numeric_feature_cols_for_plot = df[feature_cols].select_dtypes(include=np.number).columns.tolist()

    if not numeric_feature_cols_for_plot:
        print("No numeric feature columns available for plotting. Skipping visualization section.")
    else:
        fig = plt.figure(figsize=(22, 18))
        plot_idx = 1

        # Graph 1: AS Relationship Type Distribution (Pie Chart)
        # Tooltip: "This pie chart shows the proportion of Peer-to-Peer (P2P) versus Customer-Provider (C2P) relationships in the dataset."
        if 'relationship_type' in df.columns:
            plt.subplot(3, 4, plot_idx); plot_idx+=1
            df['relationship_type'].value_counts().plot(kind='pie',
                                                       labels=['Peer-to-Peer (0)', 'Customer-Provider (1)'],
                                                       autopct='%1.1f%%', colors=['skyblue', 'lightcoral'], startangle=90)
            plt.title('AS Relationship Type Distribution', fontsize=12, fontweight='bold')
            plt.ylabel('')

        # Graph 2: AS Degree Distribution (Scatter Plot) - if 'as1_degree', 'as2_degree' exist
        # Tooltip: "Scatter plot of AS1 Degree vs AS2 Degree, colored by relationship type. Helps see if degree correlates with relationship."
        if 'as1_degree' in df.columns and 'as2_degree' in df.columns:
            plt.subplot(3, 4, plot_idx); plot_idx+=1
            scatter = plt.scatter(df['as1_degree'], df['as2_degree'],
                        c=df['relationship_type'], alpha=0.6, cmap='coolwarm', s=25)
            plt.xlabel('AS1 Degree', fontsize=10); plt.ylabel('AS2 Degree', fontsize=10)
            plt.title('AS Degree Comparison', fontsize=12, fontweight='bold')
            min_val = min(df['as1_degree'].min(), df['as2_degree'].min())
            max_val = max(df['as1_degree'].max(), df['as2_degree'].max())
            plt.plot([min_val, max_val], [min_val, max_val], 'k--', alpha=0.4, lw=1)
            plt.legend(*scatter.legend_elements(), title="Type", loc="upper right")


        # Box plots for several features
        boxplot_features = {
            'avg_path_length_diff': "Avg. Path Length Diff.",
            'prefix_ratio': "Prefix Ratio (AS1/AS2)",
            'geographic_distance': "Geo. Distance (km)",
            'common_neighbors': "# Common Neighbors",
            'community_similarity': "Community Similarity",
            'prepending_freq': "Path Prepending Freq."
        }
        # Tooltips for boxplots: "Compares distribution of [Feature Name] for P2P (0) vs C2P (1) relationships."
        for col, title_text in boxplot_features.items():
            if col in df.columns and plot_idx <= 10: # Limit plots to available space + existing correlation plot
                plt.subplot(3, 4, plot_idx); plot_idx+=1
                sns.boxplot(x='relationship_type', y=col, data=df, palette="pastel")
                plt.xticks([0, 1], ['P2P (0)', 'C2P (1)'])
                plt.xlabel('Relationship Type', fontsize=10); plt.ylabel(title_text, fontsize=10)
                plt.title(f'{title_text} by Relationship', fontsize=12, fontweight='bold')


        # Graph 10 (or last available slot): Feature Correlation with Target
        # Tooltip: "Bar chart showing correlation of each feature with the relationship type. Indicates linear relationships."
        if plot_idx <= 11: # Slot for correlation plot (originally subplot 3,4,10)
             plt.subplot(3, 4, min(plot_idx, 10)); # Use slot 10 or next available
        else: # If other plots took up to 10, use the next one if available
             if plot_idx <=12: plt.subplot(3,4, plot_idx)

        if 'relationship_type' in df.columns and len(numeric_feature_cols_for_plot) > 0:
            correlations = df[numeric_feature_cols_for_plot].corrwith(df['relationship_type']).sort_values(ascending=False)
            correlations.plot(kind='barh', color='teal', ax=plt.gca())
            plt.xlabel('Correlation with Relationship Type', fontsize=10)
            plt.title('Feature Correlations (Linear)', fontsize=12, fontweight='bold')
        else:
            plt.text(0.5, 0.5, "Not enough data for correlation plot", ha='center', va='center')
            plt.title('Feature Correlations (Linear)', fontsize=12, fontweight='bold')


        plt.suptitle('BGP Network Features - Exploratory Analysis', fontsize=16, fontweight='bold', y=1.00)
        plt.tight_layout(rect=[0, 0, 1, 0.98])
        plt.show()

        # ========== SECTION 5: CORRELATION HEATMAP ==========
        # Tooltip: "Heatmap of pairwise correlations between all features. Helps identify multicollinearity."
        plt.figure(figsize=(14, 12))
        if len(numeric_feature_cols_for_plot) > 1:
            cols_for_corr = numeric_feature_cols_for_plot + ['relationship_type']
            df_corr = df[cols_for_corr].dropna()
            if not df_corr.empty and df_corr.shape[1] > 1:
                correlation_matrix = df_corr.corr()
                mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))
                sns.heatmap(correlation_matrix, mask=mask, annot=True, cmap='coolwarm',
                            center=0, fmt='.2f', square=True, linewidths=0.5, annot_kws={"size": 8})
                plt.title('Feature Correlation Heatmap', fontsize=14, fontweight='bold')
                plt.xticks(rotation=45, ha="right", fontsize=9); plt.yticks(fontsize=9)
                plt.tight_layout()
                plt.show()
            else: print("\nNot enough valid data to generate correlation heatmap after dropping NaNs.")
        else: print("\nInsufficient numeric features for correlation heatmap.")

    # ========== SECTION 6: STATISTICAL SUMMARY ==========
    print("\n4. STATISTICAL ANALYSIS BY RELATIONSHIP TYPE") # Original section numbering was off
    print("-" * 50)
    if not df.empty and 'relationship_type' in df.columns and len(feature_cols) > 0:
        summary_stats_list = []
        for col in feature_cols: # Use all defined feature_cols
            if col in df.columns and df[col].dtype in [np.float64, np.int64]: # Ensure numeric for describe
                p2p_series = df[df['relationship_type'] == 0][col]
                c2p_series = df[df['relationship_type'] == 1][col]
                summary_stats_list.append({
                    'Feature': col,
                    'P2P Mean': p2p_series.mean(), 'P2P Median': p2p_series.median(), 'P2P Std': p2p_series.std(),
                    'C2P Mean': c2p_series.mean(), 'C2P Median': c2p_series.median(), 'C2P Std': c2p_series.std(),
                    'Mean Diff (C2P-P2P)': c2p_series.mean() - p2p_series.mean()
                })
        if summary_stats_list:
            summary_df_stats = pd.DataFrame(summary_stats_list).set_index('Feature')
            print("\nFeature Statistics by Relationship Type (DataFrame Display):")
            display(summary_df_stats) # Removed .style.format to avoid potential AttributeError
        else: print("No numeric features available to generate summary statistics DataFrame.")
    else: print("DataFrame empty or necessary columns missing for statistical summary.")


    # ========== SECTION 7: DATA PREPARATION FOR MODELING ==========
    print("\n5. PREPARING DATA FOR LOGISTIC REGRESSION") # Original section numbering
    print("-" * 50)

    model_ready = False
    if not df.empty and 'relationship_type' in df.columns and len(feature_cols) > 0:
        # Drop rows with NaN in features or target before modeling
        df_model = df[feature_cols + ['relationship_type']].dropna()
        if not df_model.empty:
            X = df_model[feature_cols]
            y = df_model['relationship_type'].astype(int)

            if len(X) > 10 and len(y.unique()) > 1: # Basic check for sufficient data
                X_train, X_test, y_train, y_test = train_test_split(
                    X, y, test_size=0.25, random_state=42, stratify=y # Increased test_size slightly
                )
                print(f"Training set size: {X_train.shape}, Test set size: {X_test.shape}")
                print(f"\nClass distribution in training set (after dropna):")
                display((y_train.value_counts(normalize=True) * 100).to_frame(name='Percentage'))

                scaler = StandardScaler()
                X_train_scaled = scaler.fit_transform(X_train)
                X_test_scaled = scaler.transform(X_test)
                model_ready = True
            else: print("Insufficient data or classes after NaN removal for modeling.")
        else: print("DataFrame became empty after removing NaNs. Cannot prepare for modeling.")
    else: print("Cannot prepare data for modeling: DataFrame or essential columns are empty/missing.")


    # ========== SECTION 8: QUICK LOGISTIC REGRESSION PREVIEW ==========
    if model_ready:
        print("\n6. LOGISTIC REGRESSION MODEL PREVIEW") # Original section numbering
        print("-" * 50)
        lr_model = LogisticRegression(random_state=42, max_iter=1000, class_weight='balanced', solver='liblinear')
        lr_model.fit(X_train_scaled, y_train)

        y_pred = lr_model.predict(X_test_scaled)
        y_pred_proba = lr_model.predict_proba(X_test_scaled)[:, 1]

        print("\nModel Performance Metrics:")
        target_names_map = {0: 'Peer-to-Peer (0)', 1: 'Customer-Provider (1)'}
        actual_target_names = [target_names_map[val] for val in sorted(y_test.unique())]
        print(classification_report(y_test, y_pred, target_names=actual_target_names))
        print("\nROC AUC Score:", roc_auc_score(y_test, y_pred_proba))

        if hasattr(lr_model, 'coef_'):
            feature_importance = pd.DataFrame({
                'Feature': X_train.columns, # Use columns from X_train (which are feature_cols after potential dropna)
                'Coefficient': lr_model.coef_[0]
            }).sort_values(by='Coefficient', key=abs, ascending=False)
            print("\nTop Features (based on coefficient magnitude):")
            display(feature_importance.head(10))

            plt.figure(figsize=(10, min(6, len(feature_importance)*0.5) )) # Adjust height
            sns.barplot(x='Coefficient', y='Feature', data=feature_importance.head(15), palette='viridis_r')
            plt.title('Top Feature Importances (Logistic Regression)', fontsize=14, fontweight='bold')
            plt.tight_layout(); plt.show()
    else: print("\nSkipping Logistic Regression Model Preview as data preparation failed or was insufficient.")

    # ========== SECTION 9: SAVE PREPARED DATA ==========
    print("\n7. SAVING PROCESSED DATA") # Original section numbering
    print("-" * 50)
    if not df.empty:
        output_file = 'bgp_as_relationships_processed_final.csv'
        df.to_csv(output_file, index=False)
        print(f"Processed data saved to: {output_file}")
        if feature_cols:
            with open('feature_names_final.txt', 'w') as f: f.write('\n'.join(feature_cols))
            print("Feature names saved to: feature_names_final.txt")
    else: print("DataFrame is empty. No data saved.")

    print("\n" + "=" * 80)
    print("ANALYSIS PIPELINE COMPLETE")
    print(f"Script finished at: {datetime.now()}")
    print("=" * 80)

    if not df.empty and 'relationship_type' in df.columns:
        print("\nFINAL DATASET SUMMARY (used for modeling if successful):")
        source_df = df_model if model_ready else df # Use df_model if modeling happened, else original df
        print(f"Total AS pairs in summary: {len(source_df)}")
        print(f"Number of features: {len(feature_cols)}")
        peer_count = (source_df['relationship_type'] == 0).sum()
        customer_count = (source_df['relationship_type'] == 1).sum()
        total_count = len(source_df)
        if total_count > 0:
            print(f"Peer-to-Peer relationships: {peer_count} ({peer_count / total_count * 100:.1f}%)")
            print(f"Customer-Provider relationships: {customer_count} ({customer_count / total_count * 100:.1f}%)")
    else: print("\nFINAL DATASET SUMMARY: DataFrame was empty or relationship_type missing.")
```
